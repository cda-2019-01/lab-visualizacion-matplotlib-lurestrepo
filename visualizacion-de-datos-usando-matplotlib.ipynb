## Punto 1

Construya una gr치fica similar a la siguiente usando el archivo `indicadores-mundiales.csv`. Se grafica la suma del porcetaje de la poblaci칩n por regiones.

![act2](images/suma-pob-region-edades-2.png)


```python
df=pd.read_csv(open('indicadores-mundiales.csv'), delimiter=',')
```

```python
copia = df.copy()
(df.loc[df['Region'] == "africa", "Region"]) = "Africa"
```

```python
aux = (df.groupby('Region')['Poblacion 0-14','Poblacion 15-64','Poblacion 65+'].sum() / 100).sort_values(by='Region')
aux
```

```python
auxT = aux.T
auxT
```

```python
fig, axs = plt.subplots(1, 6, sharex='col', sharey='row', figsize=(14, 10));
plt.subplots_adjust(wspace = 0, hspace=0.1)
plt.style.use('default')
colors = ['blue', 'orange', 'green']
for i, colname in enumerate(auxT.columns):
    axs[i].scatter(list(auxT.index), auxT[colname], s=200, edgecolors=colors, linewidths=1, 
                       facecolors='none')
    axs[i].set_title(colname)
    axs[0].set_ylabel('Valor')
    plt.style.use('default')
    for tick in axs[i].get_xticklabels():
        tick.set_rotation(90)
        plt.subplots_adjust(left =0.05)
       
```

```python
df=pd.read_csv(open('indicadores-mundiales.csv'), delimiter=',')
copia = df.copy()
(df.loc[df['Region'] == "africa", "Region"]) = "Africa"
soloano=[]
soloano = [ z.split('-')[0] for z in df.Ano]
df['soloano']=soloano
```
